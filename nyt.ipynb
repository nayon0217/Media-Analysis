```python
current_date = start_date
while current_date <= end_date:
    date = current_date.strftime('%Y/%m/%d')
    URL = f"https://www.nytimes.com/issue/todayspaper/{date}/todays-new-york-times"
    driver.get(url = URL)
    time.sleep(0.5)
    print(date)

    # front page
    articles = driver.find_elements(By.CLASS_NAME,  "story-body")
    for a in articles:
        url = a.find_element(By.TAG_NAME, 'a').get_attribute('href')
        title = a.find_element(By.TAG_NAME,'a').text
        category = 'front-page'

        urls.append(url)
        titles.append(title)
        categories_.append(category)
        dates.append(date)


    front_page_articles = driver.find_element(By.CLASS_NAME, "css-12y5jls").find_elements(By.CLASS_NAME, "css-141drxa")
    for a in front_page_articles:
        url = a.find_element(By.TAG_NAME, 'a').get_attribute('href')
        title = a.find_element(By.TAG_NAME, 'a').find_element(By.TAG_NAME, 'h2').text
        category = 'front-page'

        urls.append(url)
        titles.append(title)
        categories_.append(category)
        dates.append(date)


    # international

    categories = driver.find_elements(By.CLASS_NAME, "css-12etkvn")
    categories = categories[1:]
    for i in range(len(categories)):
        c = categories[i]
        category = c.find_elements(By.TAG_NAME, 'a')[0].get_attribute('name')
        articles = c.find_elements(By.TAG_NAME, 'li')
        for a in articles:
            url = a.find_element(By.TAG_NAME, 'a').get_attribute('href')
            title = a.find_element(By.TAG_NAME, 'a').find_element(By.TAG_NAME, 'h2').text

            urls.append(url)
            titles.append(title)
            categories_.append(category)
            dates.append(date)

    current_date += timedelta(days=1)
```
