```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
```

```python
file_path = 'cnn_low-income_links.csv'
df = pd.read_csv(file_path)
links = []
for index, row in df.iterrows():
    # Convert the row to a list and then print it
    links.append(list(row)[1])

```

```python
len(links)
```

```python
urls = []
titles = []
timestamps = []
contents = []
video_urls = []
underscored_urls = []

from bs4 import BeautifulSoup
import requests

for i in range(2,998):
    url = links[i]
    print(url)

    if '/video/' in url or '/videos/' in url:
        video_urls.append(url)

    elif '/cnn-underscored/' in url:
        underscored_urls.append(url) 
    
    elif i == 206:
        pass

    else:
        response = requests.get(url)
        html_content = response.text
        soup = BeautifulSoup(html_content, 'html.parser')
        title = soup.find('h1', class_="headline__text").text.strip()
        timestamp = soup.find_all('div', class_="timestamp")[0].text.strip()
        textbox = soup.find('div', class_ = "article__content-container").find_all('p', class_ = "paragraph")
        content = ""
        for text_ in textbox:
            content = content + text_.text.strip()

        urls.append(url)
        titles.append(title)
        timestamps.append(timestamp)
        contents.append(content)

```

```python
len()
```

```python
df = pd.DataFrame({
    "url":urls,
    "title":titles,
    "content":contents,
    "timestamp":timestamps
})

df.to_csv("cnn_low-income_articles.csv")


df2 = pd.DataFrame({
    "url":video_urls
})

df2.to_csv("cnn_low-income_videos.csv")


df3 = pd.DataFrame({
    "url":underscored_urls
})

df3.to_csv("cnn_low-income_underscored.csv")
```
